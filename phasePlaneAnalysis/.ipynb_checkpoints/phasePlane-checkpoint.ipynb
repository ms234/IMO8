# Phase Plane

```python
import numpy as np
import pandas as pd
import scipy.integrate
import math
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# from matplotlib import animation, rc
# from IPython.display import HTML
# import time

# Format plot
sns.set(style="white", 
        rc={'figure.figsize':(12,6), 
            'lines.linewidth': 3})
```

```python
def modelEqs(uVec,t,parsVec,drugVec,tOnTxVec):
    # Unpack parameters
    T, E, S = uVec
    lambda1, lambda2, lambda3, alpha, K, eta, sigma, rho, gamma, mu, delta, rho, epsilon = parsVec
    Rmtor, Rmtki, Rnivo = drugVec
    dudtVec = np.zeros_like(uVec)
    # Compute the current sensitivity to each treatment
    beta = math.exp(-lambda1*(t-tOnTxVec[0]))
    phi = math.exp(-((t-tOnTxVec[1])/lambda2)**2)
    psi = 0.5*math.exp(-((t-tOnTxVec[2])/lambda3)**2)
    # Compute the derivatives
    dudtVec[0] = (alpha - beta*Rmtor)*T*(1 - (T/K)**(1-phi*Rmtki)) - eta*E*T*(1 - (S/(S+E)))
    dudtVec[1] = sigma + ((rho + psi*Rnivo)*E*T/(gamma + T)) - mu*E*T - delta*E
    dudtVec[2] = rho*S*T/(gamma + T) - epsilon*S;
    return dudtVec
```

```python
# Parameters
T0 = 1.5E5#1.5E9  # Initial tumour cell numbers
E0 = 2E6 # Initial effector cell numbers
S0 = 3*E0  # Initial suppressor cell numbers

# Treatment parmeters
lambda1 = 0.18
lambda2 = 10 
lambda3 = 0.005

# Model parameters
alpha = 0.18 * 0.2 # Tumour growth rate
K = 1E13 # Tumour carrying capacity
eta   = 1.1E-8 # Lymphocyte kill rate ; %1.101E-7; 
sigma = 1.3E4 # Lymphocyte influx rate
rho = 0.1245 # Lymphocyte recruitement rate
gamma = 2.019E7 # Half life of recruitment 
mu = 3.422E-10 # Lymphocyte exhaustion rate % rho/(gamma+K) *10; %
delta = 0.0412 # Lymphoocyte death rate %0.04; %0.01412;
epsilon = 0.9*rho*K/(gamma+K) # Suppressor rate death rate

tEnd = 5*365 # End time
tDrugOn = 30 # Start of treatment


# Initialise variables for solution
drugVec = [0,0,0]
tOnTxVec = [0,0,0]
parsVec = [lambda1, lambda2, lambda3, alpha, K, eta, sigma, rho, gamma, mu, delta, rho, epsilon]

# Solve the equation
tEvalVec = np.linspace(0,tEnd,int(1e3))
solverObj = scipy.integrate.solve_ivp(lambda t,y: modelEqs(y,t,parsVec,drugVec,tOnTxVec), 
                                   t_span=(0,tEnd), y0=[T0,E0,S0],
                                   method='BDF', t_eval=tEvalVec,
                                   rtol=1e-6, atol=1e-9)
```

```python
plt.figure(figsize=(12,12))
# Plot the tumour
plt.subplot(2,2,1)
plt.plot(tEvalVec,np.log10(solverObj.y[0,:]))
plt.title('Tumour Cells')
plt.xlabel('Time in Days')
plt.ylabel('Number of Cells')
# Plot the effector cells
plt.subplot(2,2,2)
plt.plot(tEvalVec,solverObj.y[1,:])
plt.title('Effector Cells')
plt.xlabel('Time in Days')
plt.ylabel('Number of Cells')
# Plot the suppressor cells
plt.subplot(2,2,3)
plt.plot(tEvalVec,solverObj.y[2,:])
plt.title('Suppressor Cells')
plt.xlabel('Time in Days')
plt.ylabel('Number of Cells')
# Plot the NLR
plt.subplot(2,2,4)
plt.plot(tEvalVec,solverObj.y[2,:]/solverObj.y[1,:])
plt.title('NLR')
plt.xlabel('Time in Days')
plt.ylabel('NLR')

plt.tight_layout()
```

## Plot the phase plane

```python
TumourValsVec = np.linspace(1e-5, K, 25)
EffValsVec = np.linspace(1e-5, 8e6, 25)
SuppValsVec = S0 #np.linspace(0, 1e13, 100)

TumourValsMat, EffValsMat, SuppValsMat = np.meshgrid(TumourValsVec, EffValsVec, SuppValsVec)

t = 0

dTdt, dEdt, dSdt = np.zeros(TumourValsMat.shape), np.zeros(EffValsMat.shape), np.zeros(SuppValsMat.shape)

NI, NJ, NK = TumourValsMat.shape

for i in range(NI):
    for j in range(NJ):
        for k in range(NK):
            T = TumourValsMat[i, j, k]
            E = EffValsMat[i, j, k]
            S = SuppValsMat[i, j, k]
            yprime = modelEqs([T,E,S],0,parsVec,drugVec,tOnTxVec)
            dTdt[i,j] = yprime[0]
            dEdt[i,j] = yprime[1]
            dSdt[i,j] = yprime[2]

# Plot the phase diagram
plt.figure(figsize=(12,12))
q = plt.quiver(EffValsMat, TumourValsMat, dEdt[:,:,0], dTdt[:,:,0], color='b', pivot='mid')

# Plot the steady states
# State for tumour escape
ss_Escape = scipy.optimize.root(modelEqs,[K,0.1,0.1],args=(0,parsVec,drugVec,tOnTxVec)).x
ss_Control = scipy.optimize.root(modelEqs,[0.1,sigma/delta,0.1],args=(0,parsVec,drugVec,tOnTxVec)).x
plt.scatter(ss_Escape[1],ss_Escape[0])
plt.scatter(ss_Control[1],ss_Control[0])
plt.plot(solverObj.y[1,:],solverObj.y[0,:])

# Decorate
plt.xlabel('Effector')
plt.ylabel('Tumour')
```
